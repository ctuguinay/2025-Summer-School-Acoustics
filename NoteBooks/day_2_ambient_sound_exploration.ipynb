# Playing with OOI Data
Let's look at some sections of OOI hydrophone data and see if we can identify the different sounds

## Source Identification

### Segment 1
**Location**
- Central Caldera, low frequency hydrophone (AXCC1) </br>

**Time Range**
- Time range: 2018-12-15 T00:00:00 to 2018-12-15 T00:30:00

### Segment 2
**Location**
- Axial Base, low frequency hydrophone (AXBA1)

**Time Range**
- 2015-04-24 T00:00:00 to 2015-04-24 T12:00:00

### Segment 3

**Locations**
- Axial Base, broadband hydrophone (LJ03A)
- Central Caldera, low frequency hydrophone (AXCC1)

**Time Range**
- 2019-07-29 T07:00:00 to 2019-07-29 T07:01:00


```python
hdata = ooipy.get_acoustic_data(datetime(2019,7,29,7), datetime(2019,7,29,7,1), 'LJ03A', jupyter_hub=True)
```

```python
hdata_norm = hdata.data - np.mean(hdata.data)
hdata_norm = hdata_norm / np.max(hdata_norm)
```

```python
Audio(hdata, rate=64000)
```

```python
spec = hdata.compute_spectrogram()
```

```python
spec.plot(x='time', vmax=110, vmin=50)
plt.xlim([pd.Timestamp('2019-07-29 T07:00:10'), pd.Timestamp('2019-07-29 T07:00:15')])
```

```python
oa.slice_lf_hydrophones(hyds['AXBA1'], pd.Timestamp(2019,7,29,), pd.Timestamp(2019,7,30), include_coord=True).plot()
```

### Segment 4
**Locations**
- Central Caldera, low frequency hydrophone (AXCC1)

**Time Range**
- 2019-07-01 T00:00:00 to 2019-09-01 T00:00:00

```python
import xrsignal as xrs
import seaborn as sns
```

```python
hslice = oa.slice_lf_hydrophones(hyds, pd.Timestamp('2019-07-01'), pd.Timestamp('2019-08-15'), include_coord=False)
```

```python
specs = xrs.welch(hslice['AXCC1'],dim='time',dB=True, nperseg=2048)
```

```python
specs = specs.compute()
```

```python
fig = plt.figure(figsize=(10,3))
(specs + 120).plot(vmax=130, vmin=50, cmap='rocket')
```
